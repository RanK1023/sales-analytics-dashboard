```python
date_cols=['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date']
for col in date_cols:
    df[col]=pd.to_datetime(df[col], errors='coerce')
```

```python
df['undelivered_order_customer_date']=df['order_delivered_customer_date'].isna()
df['undelivered_order_customer_date'].sum()
```

```python
df
```

```python
df['delivery_time_days']=(df['order_delivered_customer_date']- df['order_purchase_timestamp']).dt.days
```

```python
df['delivery_time_days'].head()
```

```python
df['delivery_delay_days']= ( df['order_delivered_customer_date']-df['order_estimated_delivery_date']).dt.days
df['delivery_delay_days'].head()
```

```python
df['is_late']=(df['delivery_delay_days']>0)
df['is_late'].head()
```

```python
df['seller_handling_days']=(df['order_delivered_carrier_date']-df['order_approved_at']).dt.days
df['seller_handling_days'].head()
```

```python
df['carrier_transit_days']=(df['order_delivered_customer_date']-df['order_delivered_carrier_date']).dt.days
df['carrier_transit_days'].head()
```
